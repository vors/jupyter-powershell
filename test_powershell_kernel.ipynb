# Baseline Notebook Tests for PowerShell Kernel

```powershell
# Test for Single line output

Write-Host Hello
# Unit Test 1 - Expected Result: Hello (once)
```

```powershell
# Test for Interactive Streaming Support

For ($i=0; $i -le 4; $i++) {
    Write-Host Hello
    Start-Sleep 1
}
# Unit Test 2 - Expected Result: Hello (x5)
```

```powershell
# Test for Large String Streaming Support

$x = "Start123abc" * 1000 # replicates string 1000 times
For ($i=0; $i -le 1; $i++) {
    Write-Host $x 
    Start-Sleep 1
}
# Unit Test 3 - Expected Result: Start123abcEnd... (outputted twice seperated by new line/space)
```
